```python
results = []

cols = df.select_dtypes(include=['int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 3*iqr
  upper_bound = q3 + 3*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

```python
def recommend_similar_Heroes(cluster, df_fixdata, n_recommendations=10):
    # Filter DataFrame berdasarkan cluster yang diinput
    filtered_heroes = df_fixdata[df_fixdata['cluster'] == cluster]

    if cluster == 1 and not filtered_heroes.empty:
        # Ambil nilai rata-rata total pick dan total win
        avg_T_Picked = filtered_heroes['T_Picked'].mean()
        avg_T_Wins = filtered_heroes['T_Wins'].mean()

        # Filter hero berdasarkan total pick dan total win yang mirip dengan rata-rata
        recommended_heroes = filtered_heroes[
            (filtered_heroes['T_Picked'] >= avg_T_Picked) &
            (filtered_heroes['T_Wins'] >= avg_T_Wins)
        ]

        # Ambil n hero dengan kluster yang sama secara acak
        recommended_heroes = recommended_heroes.sample(n=min(n_recommendations, len(recommended_heroes)), random_state=42)

        # Dapatkan nama hero yang direkomendasikan
        recommended_titles = recommended_heroes[['Hero', 'Roles']].values.tolist()  # Perbaikan di sini

    else:
        recommended_titles = []

    return recommended_titles

# Contoh penggunaan
input_cluster = 1  # Ubah sesuai dengan cluster yang diinginkan
recommended_heroes = recommend_similar_Heroes(input_cluster, df_fixdata)
if recommended_heroes:
    print("Rekomendasi Hero untuk dipick berdasarkan Tingkat kemenangan dan Total Picknya:".format(input_cluster))
    for hero, roles in recommended_heroes:  # Perbaikan di sini
        print("- {} ({})".format(hero, roles))  # Perbaikan di sini
else:
    print("Tidak ada rekomendasi untuk cluster {}.".format(input_cluster))
```

```python
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import calinski_harabasz_score, davies_bouldin_score, silhouette_score

# Misalkan df_fixdata adalah dataframe yang berisi data yang sudah diencode

# Inisialisasi model K-Means dengan jumlah klaster yang diinginkan
num_clusters = 2
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(df_fixdata.drop(['Hero', 'Roles'], axis=1))

# Hitung skor untuk setiap model klaster
calinski_harabasz_scores = []
davies_bouldin_scores = []
silhouette_scores = []

# Hitung skor untuk K-Means Clustering
kmeans_calinski_harabasz = calinski_harabasz_score(df_fixdata.drop(['Hero', 'Roles'], axis=1), kmeans.labels_)
kmeans_davies_bouldin = davies_bouldin_score(df_fixdata.drop(['Hero', 'Roles'], axis=1), kmeans.labels_)
kmeans_silhouette = silhouette_score(df_fixdata.drop(['Hero', 'Roles'], axis=1), kmeans.labels_)

calinski_harabasz_scores.append(kmeans_calinski_harabasz)
davies_bouldin_scores.append(kmeans_davies_bouldin)
silhouette_scores.append(kmeans_silhouette)

# Inisialisasi model Hierarchical Agglomerative Clustering (HAC) dengan jumlah klaster yang diinginkan
hac = AgglomerativeClustering(n_clusters=num_clusters)
hac.fit(df_fixdata.drop(['Hero', 'Roles'], axis=1))

# Hitung skor untuk Hierarchical Agglomerative Clustering (HAC)
hac_calinski_harabasz = calinski_harabasz_score(df_fixdata.drop(['Hero', 'Roles'], axis=1), hac.labels_)
hac_davies_bouldin = davies_bouldin_score(df_fixdata.drop(['Hero', 'Roles'], axis=1), hac.labels_)
hac_silhouette = silhouette_score(df_fixdata.drop(['Hero', 'Roles'], axis=1), hac.labels_)

calinski_harabasz_scores.append(hac_calinski_harabasz)
davies_bouldin_scores.append(hac_davies_bouldin)
silhouette_scores.append(hac_silhouette)

# Susun ke dalam dataframe
metric = pd.DataFrame({
    'Silhouette': silhouette_scores,
    'Calinski Harabasz': calinski_harabasz_scores,
    'Davies Bouldin': davies_bouldin_scores
}, index=['KMeans', 'Hierarchical'])

metric
```

Nilai yang lebih tinggi menunjukkan bahwa objek lebih baik cocok dengan klaster mereka sendiri dan tidak cocok dengan klaster tetangga.

Calinski Harabasz: Calinski-Harabasz Score adalah rasio antara dispersi dalam klaster dan dispersi antara klaster. Skor ini memberikan ukuran seberapa baik klaster tersebut dipisahkan dari satu sama lain. Nilai yang lebih tinggi menunjukkan klaster yang lebih terpisah dan lebih kompak.

Davies Bouldin: Davies-Bouldin Score merupakan rasio dari jarak antara klaster dengan kepadatan dalam klaster. Skor ini memberikan ukuran seberapa baik sebuah klaster dipisahkan satu sama lain. Semakin rendah nilainya, semakin baik klaster tersebut dipisahkan.

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Tuning K-Means
best_silhouette_score = -1
best_calinski_harabasz_score = -1
best_davies_bouldin_score = float('inf')
best_kmeans_model = None
best_k = 0

for k in range(2, 4):  # Coba jumlah cluster dari 2 hingga 4
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_fixdata.drop(['Hero','Roles'], axis=1))
    labels = kmeans.labels_
    silhouette_avg = silhouette_score(df_fixdata.drop(['Hero','Roles'], axis=1), labels)
    calinski_harabasz_avg = calinski_harabasz_score(df_fixdata.drop(['Hero','Roles'], axis=1), labels)
    davies_bouldin_avg = davies_bouldin_score(df_fixdata.drop(['Hero','Roles'], axis=1), labels)

    if silhouette_avg > best_silhouette_score:
        best_silhouette_score = silhouette_avg
        best_kmeans_model = kmeans
        best_k = k
    if calinski_harabasz_avg > best_calinski_harabasz_score:
        best_calinski_harabasz_score = calinski_harabasz_avg
    if davies_bouldin_avg < best_davies_bouldin_score:
        best_davies_bouldin_score = davies_bouldin_avg

print("Parameter terbaik untuk K-Means:")
print("Jumlah Cluster Terbaik:", best_k)
print("Silhouette Score Terbaik:", best_silhouette_score)
print("Calinski Harabasz Score Terbaik:", best_calinski_harabasz_score)
print("Davies Bouldin Score Terbaik:", best_davies_bouldin_score)

# Tuning Agglomerative Clustering
best_silhouette_score = -1
best_calinski_harabasz_score = -1
best_davies_bouldin_score = float('inf')
best_agg_model = None
best_n_clusters = 0

for n_clusters in range(2, 4):  # Coba jumlah cluster dari 2 hingga 4
    agg = AgglomerativeClustering(n_clusters=n_clusters)
    labels = agg.fit_predict(df_fixdata.drop(['Hero','Roles'], axis=1))   # Menggunakan data train untuk fitting dan prediksi
    silhouette_avg = silhouette_score(df_fixdata.drop(['Hero','Roles'], axis=1), labels)
    calinski_harabasz_avg = calinski_harabasz_score(df_fixdata.drop(['Hero','Roles'], axis=1), labels)
    davies_bouldin_avg = davies_bouldin_score(df_fixdata.drop(['Hero','Roles'], axis=1), labels)

    if silhouette_avg > best_silhouette_score:
        best_silhouette_score = silhouette_avg
        best_agg_model = agg
        best_n_clusters = n_clusters
    if calinski_harabasz_avg > best_calinski_harabasz_score:
        best_calinski_harabasz_score = calinski_harabasz_avg
    if davies_bouldin_avg < best_davies_bouldin_score:
        best_davies_bouldin_score = davies_bouldin_avg

print("\nParameter terbaik untuk Agglomerative Clustering:")
print("Jumlah Cluster Terbaik:", best_n_clusters)
print("Silhouette Score Terbaik:", best_silhouette_score)
print("Calinski Harabasz Score Terbaik:", best_calinski_harabasz_score)
print("Davies Bouldin Score Terbaik:", best_davies_bouldin_score)
```
